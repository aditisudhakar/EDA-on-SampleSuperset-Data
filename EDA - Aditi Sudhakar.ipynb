### Visualising the dataset

```python
print(df['Sub-Category'].value_counts())
plt.figure(figsize=(12,6))
sns.countplot(x=df['Sub-Category'])
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.kdeplot(df['Sales'],color='Green',label='Sales',shade=True,bw=25)
sns.kdeplot(df['Profit'],color='Orange',label='Profit',shade=True,bw=25)
plt.xlim([-100,1000])
plt.legend()
```

From this plot we see that the overall profit is higher than the sales, but there are definitely areas where the profit could be improved. 

```python
plt.figure(figsize = (10,6))
sns.lineplot('Discount', 'Profit', data = df, color = 'blue', label= 'Discount')
plt.legend()
```

### Heatmap to show Correlation

```python
df.corr()
```

```python
fig,axes = plt.subplots(1,1,figsize=(10,6))
sns.heatmap(df.corr(), annot= True)
plt.show()
```

Observations:
* Postal code is not significantly correlated to any field.
* Sales is negatively correlated to discount and has moderate positive correlation to sales
* Quantity is not significantly correlated to any field.
* Discount is negatively correlated to sales.

### Profit, Discount and Sales by Segment

```python
df['Segment'].value_counts()
```

```python
df_segment= df.groupby(['Segment'])[['Sales', 'Discount', 'Profit']].mean()
df_segment
```

```python
df_segment.plot.pie(subplots=True, 
                    autopct='%1.1f%%',
                    figsize=(18, 20),
                    startangle=90,     # start angle 90Â° (Africa)
                    shadow=True,
                    labels = df_segment.index)
plt.title('Segment wise analysis of Sale, Discount, profit')
```

### Profit, Discount and Sales by State

```python
df['State'].value_counts().head(10)
```

```python
df_state= df.groupby(['State'])[['Sales', 'Discount', 'Profit']].mean()
df_state.head(10)
```

```python
df_state1=df_state.sort_values('Profit')

df_state1[['Profit']].plot(kind = 'bar', figsize = (15,4))
plt.title('Statewise Profit Analysis', fontsize = 20)
plt.ylabel('Profit per Sate')
plt.xlabel('States')
plt.show()
```

From this we see that the state of Vermont has the highest profits and the state of Ohio has the least profits.

```python
df_state['Sales'].plot(kind='pie',
                        figsize = (20,20),
                        autopct='%1.1f%%',
                        startangle=90,     # start angle 90Â° (Africa)
                        shadow=True)
plt.title('Statewise Analysis of Sales',fontsize=20)
```

Here, the highest number of sales is in Wyoming(11.8%) and the lowest number of sales is in South Dakota(0.8%)

```python
df_state1['Discount'].plot(kind='bar',figsize=(18,5))
plt.title('Statewise Analysis of Discount', fontsize=20)
```

Th state of Illinois has the highest discounts.

### Citywise Profit Analysis 

```python
df_city= df.groupby(['City'])[['Sales', 'Discount', 'Profit']].mean()
df_city = df_city.sort_values('Profit')
df_city.head()
```

```python
df_city['Profit'].tail(30).plot(kind='bar',figsize=(15,5),color = 'Red')
plt.title('Citywise Analysis of Sale, Discount, Profit - High Profit')
```

```python
df_city['Profit'].head(30).plot(kind='bar',figsize=(15,5),color = 'Green')
plt.title('Citywise Analysis of Sale, Discount, Profit - Low Profit')
```

Here we see that 30 cities have high profit and 30 cities have negative profit. This indicated that the profit is balanced.

### Sales, Profit and Discount by Sub-Category 

```python
df_sub_category = df.groupby(['Sub-Category'])[['Sales', 'Discount', 'Profit']].mean()
df_sub_category.head(10)
```

```python
df_sub_category.sort_values('Profit')[['Sales','Profit']].plot(kind='bar',
                                                              figsize= (10,5),
                                                              label=['Avg Sales Price($)','Profit($)'])
plt.title('Sub-Categorywise Analysis of Profit')
```

Copiers bring in the highest profits

```python
plt.figure(figsize = (15,15))
plt.pie(df_sub_category['Sales'], labels = df_sub_category.index, autopct = '%1.1f%%')
plt.title('Sub-Categorywise Analysis of Sales', fontsize = 20)
plt.legend()
plt.xticks(rotation = 90)
plt.show()
```

Copiers and machines are the most sold items.

```python
plt.figure(figsize = (15,15))
plt.pie(df_sub_category['Discount'], labels = df_sub_category.index, autopct = '%1.1f%%')
plt.title('Sub-Categorywise Analysis of Discount', fontsize = 20)
plt.legend()
plt.xticks(rotation = 90)
plt.show()
```

Binders, machines and tables are items on which there is maximum discount.
